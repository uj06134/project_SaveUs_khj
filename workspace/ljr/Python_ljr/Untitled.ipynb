## 식품의약품안전처_식품영양성분DB정보
데이터 내용: 시중에 판매되는 가공식품, 외식 메뉴, 농축산물의 상세 영양 성분(탄단지, 당류, 나트륨 등) 정보입니다. 약 5~10만 건 이상의 데이터가 있습니다.
아이디어: "건강한 대체 식품 추천 & 식품 지도"

사용자가 먹은 음식과 맛은 비슷하되, 영양 성분이 더 좋은 음식을 찾아줍니다.
분석 및 시각화 (Python):
군집화 (Clustering): 수만 가지 음식을 영양 성분 기준으로 그룹핑(K-Means)합니다. (예: '고단백 저지방 그룹', '고나트륨 위험 그룹')

시각화: 버블 차트(Bubble Chart). X축은 당류, Y축은 단백질로 두고, 각 음식(점)을 뿌립니다. 사용자가 먹은 음식의 위치를 빨간색으로 표시하여 "당신이 먹은 간식은 '고당류 위험군'에 속해 있습니다"라고 시각적으로 경고합니다.

```python
df = pd.read_csv('전국통합식품영양성분정보표준데이터.csv', encoding="utf-8")
```

```python
print(df.columns)
```

```python
core_cols = [
    "식품명",
    "데이터구분명",
    "에너지(kcal)",
    "단백질(g)",
    "지방(g)",
    "탄수화물(g)",
    "당류(g)",
    "나트륨(mg)",
    "포화지방산(g)",
    "트랜스지방산(g)",
    "콜레스테롤(mg)"
]

df = df[core_cols].copy()
```

```python
# 숫자형 변환
for col in df.columns:
    if col not in ["식품명"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```
